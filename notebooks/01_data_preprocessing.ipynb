# Phase 1 : Préparation et prétraitement des données

Dans ce notebook, nous allons charger notre jeu de données FAQ, nettoyer le texte, et construire notre vocabulaire (dictionnaire de tokens).

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.data_utils import prepareData, tensorsFromPair
import random
```

## Chargement et construction du vocabulaire

```python
filepath = '../data/faq.json'
vocab, pairs = prepareData(filepath)

print(f'Nombre de paires Question/Réponse : {len(pairs)}')
print(f'Taille du vocabulaire : {vocab.n_words} mots')

print("\nExemple de paire :")
print(random.choice(pairs))
```

## Conversion en Tenseurs (Padding inclus)

```python
MAX_LENGTH = 15
pair = random.choice(pairs)
input_tensor, target_tensor = tensorsFromPair(pair, vocab, max_length=MAX_LENGTH)

print("Phrase d'entrée :", pair[0])
print("Tenseur d'entrée :", input_tensor.view(-1).tolist())
print("\nPhrase cible :", pair[1])
print("Tenseur cible :", target_tensor.view(-1).tolist())
```
