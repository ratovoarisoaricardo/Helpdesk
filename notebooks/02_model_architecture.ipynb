# Phase 2 : Architecture Seq2Seq

Dans ce notebook, nous instancions notre modèle (Encodeur, Attention, Décodeur) et vérifions les dimensions des tenseurs générés.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import torch
from src.model import EncoderRNN, AttnDecoderRNN
from src.data_utils import prepareData, tensorsFromPair
import random
```

```python
filepath = '../data/faq.json'
vocab, pairs = prepareData(filepath)

hidden_size = 128
batch_size = 1
max_length = 15

encoder = EncoderRNN(vocab.n_words, hidden_size)
decoder = AttnDecoderRNN(hidden_size, vocab.n_words)

print("Modèle initialisé.")
```

```python
# Test avec un exemple aléatoire
pair = random.choice(pairs)
input_tensor, target_tensor = tensorsFromPair(pair, vocab, max_length)

# Ajout de la dimension du batch : shape (1, max_length)
input_tensor = input_tensor.view(1, -1)
target_tensor = target_tensor.view(1, -1)

print(f"Input shape: {input_tensor.shape}")

encoder_outputs, encoder_hidden = encoder(input_tensor)
print(f"Encoder outputs shape: {encoder_outputs.shape}")
print(f"Encoder hidden shape: {encoder_hidden.shape}")

decoder_outputs, decoder_hidden, attentions = decoder(encoder_outputs, encoder_hidden, target_tensor)
print(f"Decoder outputs shape: {decoder_outputs.shape}")
print(f"Attention weights shape: {attentions.shape}")
```
